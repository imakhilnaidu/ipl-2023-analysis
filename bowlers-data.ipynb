```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
url = "https://www.espncricinfo.com/records/tournament/averages-bowling/indian-premier-league-2023-15129"

headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/42.0.2311.135 Safari/537.36 Edge/12.246"}

page = requests.get(url=url, headers=headers)
soup = BeautifulSoup(page.content, "html.parser")
```

```python
bowlers_list = []
bowlers = soup.select("table > tbody > tr")

for i in range(1, len(bowlers)+1):
    b = {}
    
    if i % 2 != 0:
        b["player"] = soup.select_one(f"table > tbody > tr:nth-child({i}) > td:nth-child(1)").text
        b["matches_played"] = soup.select_one(f"table > tbody > tr:nth-child({i}) > td:nth-child(3)").text
        b["Innings_played"] = soup.select_one(f"table > tbody > tr:nth-child({i}) > td:nth-child(4)").text
        b["balls"] = soup.select_one(f"table > tbody > tr:nth-child({i}) > td:nth-child(5)").text
        b["maidens"] = soup.select_one(f"table > tbody > tr:nth-child({i}) > td:nth-child(6)").text
        b["runs"] = soup.select_one(f"table > tbody > tr:nth-child({i}) > td:nth-child(7)").text
        b["wickets"] = soup.select_one(f"table > tbody > tr:nth-child({i}) > td:nth-child(8)").text
        b["best_innings"] = soup.select_one(f"table > tbody > tr:nth-child({i}) > td:nth-child(9)").text
        b["average"] = soup.select_one(f"table > tbody > tr:nth-child({i}) > td:nth-child(11)").text
        b["economy"] = soup.select_one(f"table > tbody > tr:nth-child({i}) > td:nth-child(12)").text
        b["strike_rate"] = soup.select_one(f"table > tbody > tr:nth-child({i}) > td:nth-child(13)").text
        b["five_wickets"] = soup.select_one(f"table > tbody > tr:nth-child({i}) > td:nth-child(14)").text
        b["ten_wickets"] = soup.select_one(f"table > tbody > tr:nth-child({i}) > td:nth-child(15)").text
        b["keeper_catches"] = soup.select_one(f"table > tbody > tr:nth-child({i}) > td:nth-child(16)").text
        b["stumpings"] = soup.select_one(f"table > tbody > tr:nth-child({i}) > td:nth-child(17)").text
        b["team"] = soup.select_one(f"table > tbody > tr:nth-child({i+1}) > td").text
    
        bowlers_list.append(b)
        print(b)
```

```python
df = pd.DataFrame(bowlers_list)
df
```

```python
df.replace("-", 0, inplace=True)
df
```
